```python
import pandas as pd
```

```python
filename = '/content/RailsToTrails_National Count Data_week(1).xlsx'
bike_df = pd.read_excel(filename)
```

```python
bike_df.head()
```

```python
#finding the values in 2020
bike_df['Week of']
```

```python
after_july_filter = bike_df['Week of']>'2020-07-27'
after_july_filter.value_counts()
```

```python
bike_df[after_july_filter].head()
```

```python
greater_than_1 = bike_df['Change 2019-2020']>1
greater_than_1.value_counts()
```

```python
bike_df[greater_than_1].head()
```

```python
df = bike_df.loc[after_july_filter & greater_than_1, :]
df
```

There are no weeks in the second half of the year where the change was greater than 100%
